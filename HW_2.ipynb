Добавлена строка plt.gca().set_aspect('equal') для верного отражения перпендикулярности прямых

```python
%matplotlib inline
 
import numpy as np
import matplotlib.pyplot as plt
import math
x = np.linspace(-100, 100, 21)
y = (-1/3) * x + 1
y2 = 3 * x + 1
plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.gca().set_aspect('equal')
```

```python
%matplotlib inline
 
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.linspace(-1, 1, 300)
y = np.linspace(-1, 1, 300)
X, Y = np.meshgrid(x,y)
F = X**2 + Y**2 - 0.6
plt.contour(X,Y,F,[0])
plt.gca().set_aspect('equal')
plt.grid()
plt.show()
```
