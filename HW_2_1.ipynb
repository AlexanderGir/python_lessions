```python
import math

def pol2cart(rho, phi):
    x = rho * math.cos(math.radians(phi))
    y = rho * math.sin(math.radians(phi))
    return(x, y)

rho=float(input("Rho: "))
phi=float(input("Phi в градусах: "))

print("x = ", rho, "y = ", phi, pol2cart(rho, phi))
```

```python
x = np.linspace(-3, 3, 300)

plt.plot(x, (np.exp(x)+ x - 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.gca().set_aspect('equal')
plt.show()

from scipy.optimize import fsolve

def a(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x * (1 - y) - 1)

x1, y1 =  fsolve(a, (2, 6))
x2, y2 =  fsolve(a, (-1, 1))
              
print (x1, y1)
print (x2, y2)
```

```python
x = np.linspace(-3, 3, 300)

plt.plot(x, (- 1/x)+2)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5) 
plt.grid()
plt.show()

from scipy.optimize import fsolve

def a(p):
    x, y = p
    return (x**2 - 1 - y, (- 1/x)+2 - y)

x1, y1 =  fsolve(a, (2, 4))
x2, y2 =  fsolve(a, (0.5, 0.5))
x3, y3 =  fsolve(a, (-1, 1))
              
print (x1, y1)
print (x2, y2)
print (x3, y3)
```
